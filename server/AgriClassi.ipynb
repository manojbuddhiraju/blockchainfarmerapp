```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def fit_model(algorithm, X_train, y_train):
  """
  Fits a classification model to the training data.
  """
  # Initialize the model
  if algorithm == "Logistic Regression":
    model = LogisticRegression()
  elif algorithm == "K-Nearest Neighbors":
    model = KNeighborsClassifier()
  elif algorithm == "Decision Tree":
    model = DecisionTreeClassifier()
  elif algorithm == "Random Forest":
    model = RandomForestClassifier()
  elif algorithm == "Naive Bayes":
    model = GaussianNB()
    
  # Fit the model to the training data
  model.fit(X_train, y_train)
  
  return model

# Load dataset
df=pd.read_csv("D:/Ph.D Work/Ph.D Work Related/Thesis Writing Work/DataSet Work/crop_prediction.csv")

# Split dataset into features (X) and labels (y)
X = df.drop("Soil_type:_Blacksoil", axis=1)
y = df["Crop:_Rice_Swarna_MTU7029"]

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Initialize a dictionary to store the results of the model comparisons
results = {}

# Define a list of classification algorithms to compare
algorithms = ["Logistic Regression", "K-Nearest Neighbors", "Decision Tree", "Random Forest", "Naive Bayes"]

# Iterate over the list of algorithms
for algorithm in algorithms:
  
  # Fit the model to the training data
  model = fit_model(algorithm, X_train, y_train)
  
  # Make predictions on the test data
  y_pred = model.predict(X_test)
  
  # Calculate and store the metrics for the model
  accuracy = accuracy_score(y_test, y_pred)
  precision = precision_score(y_test, y_pred)
  recall = recall_score(y_test, y_pred)
  f1 = f1_score(y_test, y_pred)
  
  # Store the results in the dictionary
  results[algorithm] = {"accuracy": accuracy, "precision": precision, "recall": recall, "f1": f1}

# Print the results
for algorithm, metrics in results.items():
  print(f"Results for {algorithm}:")
  print(f"Accuracy: {metrics['accuracy']}")
  print(f"Precision: {metrics['precision']}")
  print(f"Recall: {metrics['recall']}")
  print(f"F1: {metrics['f1']}")
```
